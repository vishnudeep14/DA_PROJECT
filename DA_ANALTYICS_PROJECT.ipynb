```python
import numpy
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import glob
```

```python

# setting the path for joining multiple files
files = os.path.join("C:/Users/wavef/Downloads/DA ANALYTICS PROJECT","loan_dset*.csv")

# list of merged files returned
files = glob.glob(files)

print("Resultant CSV after joining all CSV files at a particular location...");

# joining files with concat and read_csv
df = pd.concat(map(pd.read_csv, files), ignore_index=True)

df.to_csv(r'C:/Users/wavef/Downloads/DA ANALYTICS PROJECT/loan_dataset.csv', index=False)
```

```python
total=float(len(df))
gen_plot=sns.countplot(x='Gender',data=df)
for p in gen_plot.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    gen_plot.annotate(percentage, (x, y),ha='center')
plt.show()

```
